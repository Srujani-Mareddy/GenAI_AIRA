```python
import os
import time
from selenium import webdriver
import shutil
from glob import glob
from bs4 import BeautifulSoup
import PyPDF2
from collections import OrderedDict
```

```python
# Contributed by : Prasanth Bysani

base_url = "https://onlinelibrary.wiley.com/action/doSearch?SeriesKey=23719621&content=articlesChapters&sortBy=Earliest&target=default&AfterYear=2020&BeforeYear=2024"

# startPage values
start_page_range = range(0, 13)

# List of links with varying startPage values
links = [f"{base_url}&startPage={start_page}" for start_page in start_page_range]
```

```python
links
```

```python
len(links)
```

```python
#Contributed by: Prasanth Bysani


url_list=[]
for url in links:
    driver = webdriver.Chrome()
    driver.get(url)


    # BeautifulSoup to parse the HTML content of the page
    bs = BeautifulSoup(driver.page_source, "lxml")

    for h2_tag in bs.find_all('h2', {'class': 'meta__title meta__title__margin'}):

        span_tag = h2_tag.find('span', {'class': 'hlFld-Title'})
        a_tag = span_tag.find('a', {'class': 'publication_title visitable'})

        print(a_tag)
        # Extract the title and URL
        title = a_tag.text.strip()
        link = 'https://onlinelibrary.wiley.com' + a_tag.get('href')
        url_list.append(link)

        print("Title:", title)
        print("URL:", link)
    
print(url_list)

```

```python
import csv


# File path for the CSV file
csv_file_path = "urls_TA.csv"

with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['URL'])
    for url in url_list:
        writer.writerow([url])

print("CSV file created successfully:", csv_file_path)
```

```python
# Contributed by : Gayathri Tholeti

def file_download_link(href_value):
    driver = webdriver.Chrome()
    driver.get(href_value)
    time.sleep(5)
    bs = BeautifulSoup(driver.page_source, "lxml")
    link=bs.find('a',{'class':'navbar-download btn btn--cta_roundedColored'})
    download_button = link.get('href')
    download_button_link="https://onlinelibrary.wiley.com"+download_button
    return download_button_link
```

```python
# Contributed by : Nithyasree Kusakula
def move_latest_pdf(downloads_folder, destination_folder,title):

    downloads_path = os.path.expanduser(downloads_folder)
    pdf_files = glob(os.path.join(downloads_path, '*.pdf'))

    if not pdf_files:
        print("No PDF files found in the downloads folder.")
        return

    latest_pdf = max(pdf_files, key=os.path.getmtime)
    pdf_name = os.path.basename(latest_pdf)
    new_pdf_name = title

    destination_path = os.path.join(os.path.expanduser(destination_folder), new_pdf_name)
    shutil.move(latest_pdf, destination_path)

    print(f"Moved'{pdf_name}' to {destination_folder}.")
```
